```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

def plot_all(df):
    fig, axes = plt.subplots(6, 2, figsize=(18, 30))
    axes = axes.ravel()

    df_sorted = df.sort_values(by="year")
    axes[0].fill_between(df_sorted["year"], df_sorted["price"], alpha=0.5)
    axes[0].set_title("Area Plot - Цена недвижимости по годам")

    df_bar = df.groupby("district")["price"].mean().sort_values()
    df_bar.plot(kind="bar", ax=axes[1], color="skyblue")
    axes[1].set_title("Bar Plot - Средняя цена по районам")

    df_bar.plot(kind="barh", ax=axes[2], color="salmon")
    axes[2].set_title("Barh Plot - Средняя цена по районам")

    sns.boxplot(x="rooms", y="price", data=df, ax=axes[3])
    axes[3].set_title("Box Plot - Цена в зависимости от количества комнат")

    axes[4].hexbin(df["square"], df["price"], gridsize=30, cmap="Blues", mincnt=1)
    axes[4].set_title("Hexbin Plot - Цена от площади")

    sns.histplot(df["price"], bins=30, kde=False, ax=axes[5])
    axes[5].set_title("Histogram - Распределение цен")

    sns.kdeplot(df["price"], fill=True, ax=axes[6])
    axes[6].set_title("KDE Plot - Плотность распределения цен")

    sns.kdeplot(df["square"], fill=True, ax=axes[7])
    axes[7].set_title("Density Plot - Плотность распределения площади")

    df_sorted.groupby("year")["price"].mean().plot(ax=axes[8])
    axes[8].set_title("Line Plot - Средняя цена по годам")

    df_pie = df["building_type"].value_counts()
    df_pie.plot(kind="pie", ax=axes[9], autopct="%1.1f%%", startangle=90)
    axes[9].set_title("Pie Chart - Типы зданий")

    sns.scatterplot(x="square", y="price", data=df, ax=axes[10])
    axes[10].set_title("Scatter Plot - Цена vs Площадь")

    fig.delaxes(axes[11])

    plt.tight_layout()
    plt.show()

plot_all(df)
```
